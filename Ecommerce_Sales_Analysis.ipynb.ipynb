# 📊 Ecommerce Sales Analysis using Pandas & Plotly

## Project Overview
This project analyzes the Superstore dataset using Pandas for data manipulation and Plotly for interactive visualizations.

### Tools Used
- Python
- Pandas
- Plotly
- Jupyter Notebook

### Dataset
Sample Superstore Dataset

```python
import pandas as pd

import plotly .express as px
import plotly .graph_objects as go
import plotly .io as pio
import plotly .colors as colors
pio .templates.default = "plotly_white"
pio.renderers.default = "notebook_connected"
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
data.head()
```

# Converting Date Column

```python
data['Order Date'] = pd.to_datetime(
    data['Order Date'],
    errors='coerce'
)

data['Ship Date'] = pd.to_datetime(
    data['Ship Date'],
    errors='coerce'
)
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek 
```

# Monthly Sales Analysis


```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
fig=px.line(sales_by_month,
            x='Order Month',
            y='Sales',
            title='Monthly Sales Analysis')
fig.show()
```

# Sales By Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
fig = px.pie(sales_by_category,
            values='Sales',
            names='Category',
            hole=0.5,
            color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo ='percent+label')
fig.update_layout(title_text='Sales Analysis By Category',title_font=dict(size=24))

fig.show()
```

# Sales Analysis By Sub-Catergory

```python
sales_by_subcategory = data.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
fig = px.bar(sales_by_subcategory,
             x='Sub-Category',
             y='Sales',
             title= 'Sales_by_Subcategory'
            )
fig.show()
```

# Monthly Profit Analysis


```python
monthly_profit = data.groupby('Order Month')['Profit'].sum().reset_index()
```

```python
fig = px.line(monthly_profit,
             x='Order Month',
             y='Profit',
             title= 'Profit_By_Month'
            )
fig.show()
```

# Profit by Category & Sub-Category

```python
profit_by_category_n_subcategory = data.groupby(['Category','Sub-Category'])['Profit'].sum().reset_index()

```

```python
fig = px.bar(
    profit_by_category_n_subcategory,
    x='Sub-Category',
    y='Profit',
    color='Category',
    title='Profit by Category and Sub-Category'
)

fig.show()
```

# sales and profit - customer segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Profit':'sum' , 'Sales':'sum'}).reset_index()

fig = go.Figure()

# Sales Bar
fig.add_trace(
    go.Bar(
        x=sales_profit_by_segment['Segment'],
        y=sales_profit_by_segment['Sales'],
        name='Sales',
        marker_color=px.colors.qualitative.Set2[0]
    )
)

# Profit Bar
fig.add_trace(
    go.Bar(
        x=sales_profit_by_segment['Segment'],
        y=sales_profit_by_segment['Profit'],
        name='Profit',
        marker_color=px.colors.qualitative.Set2[1]
    )
)

fig.update_layout(
    title='Sales & Profit by Segment',
    xaxis_title=' Customer Segment',
    yaxis_title='Amount',
    barmode='group',
    template='plotly_white',
    title_x=0.5
)

fig.show()
```

# Sales to Profit Ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Profit':'sum' , 'Sales':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales']/ sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```

```python
fig = go.Figure()

fig.update_yaxes(
    showgrid=True,
    gridwidth=0.2,
    gridcolor='lightgray'
)

fig.update_yaxes(
    showgrid=True,
    gridwidth=0.2,
    gridcolor='lightgray'
)

fig.add_trace(
    go.Bar(
        x=sales_profit_by_segment['Segment'],
        y=sales_profit_by_segment['Sales_to_Profit_Ratio'],
        text=sales_profit_by_segment['Sales_to_Profit_Ratio'].round(2),
        textposition='outside',
        marker=dict(
            color=px.colors.qualitative.Set2,
            line=dict(color='white', width=2)
        ),
        hovertemplate='<b>%{x}</b><br>Ratio: %{y:.2f}<extra></extra>'
    )
)

fig.update_layout(
    title='<b>Sales to Profit Ratio by Customer Segment</b>',
    template='plotly_white',
    title_x=0.5,
    height=550,
    width=800,
    xaxis_title='Customer Segment',
    yaxis_title='Sales to Profit Ratio',
    yaxis=dict(
        showgrid=True,
        gridcolor='lightgray',
        zeroline=False
    ),
    xaxis=dict(
        showgrid=False
    ),
    showlegend=False
)

fig.show()
```
